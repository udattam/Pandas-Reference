# Aggregate Statistics (Groupby)

```python
df=pd.read_csv('modified.csv')
df
```

```python
df.groupby(['Type 1']).mean().sort_values('Defense',ascending=False)         #Proving that the Steel Pokemon has the highest Defense as it should be
```

```python
df.groupby(['Type 1']).mean().sort_values('Attack',ascending=False)         #Proving that the Dragon Pokemon has the highest Defense as it should be
```

```python
df.groupby(['Type 1']).mean().sort_values('HP',ascending=False)         #Proving that the Dragon Pokemon has the highest HP as it should be
```

```python
df.groupby(['Type 1']).sum()   #Sum according to groups of Type 1
```

```python
df['count']=1  #Adding a new column with the column-head count
df
```

```python
df.groupby('Type 1').count()['count'].sort_values(axis=0,ascending=False)
#Sort descending by number of Pokemons
```

```python
df.groupby(['Type 1','Type 2']).count()['count']
```

# Working with large amounts of data

```python
new_df=pd.DataFrame(columns=df.columns)
for df in pd.read_csv('modified.csv',chunksize=1000):
    results=df.groupby(['Type 1']).count()
    new_df=pd.concat([new_df,results])
new_df
```
